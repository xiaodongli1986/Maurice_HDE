##### Plottings of basic quantities

```python
def setax(ax):
    ax.set_xscale('log')
    ax.grid()
    ax.set_xlabel('$z$', fontsize=24)
    ax.set_ylabel(ax.get_ylabel(), fontsize=20)
    ax.set_xlim(0.04, 10**5.5);

omstr = '0.30'

nowfile = './z_ez_qz_Odez_Omz_Orz_RhodeToRhom_weff__mauricehde._omegam'+omstr+'.txt'
Z, Ez, Qz, Odez, Omz, Orz, RhodeToRhom, weff = Xsfromdata(np.loadtxt(nowfile), range(8))

nowfile2 = './z_ez_qz__lcdm_omegam'+omstr+'.txt'
ZLCDM, EzLCDM, QzLCDM = Xsfromdata(np.loadtxt(nowfile2), range(3))

lw=4
### Plot of H(z)/H_LCDM
fig, ax = figax()

ax.plot(Z, [Ez[row]/EzLCDM[row] for row in range(len(Ez))], lw=lw)
#ax.plot(ZLCDM, EzLCDM, label='$\\Lambda$CDM')
ax.set_ylabel('$H/H_{\\Lambda\\rm CDM}$')
setax(ax)
ax.text(0.75,0.85,'$\\Omega_m$=%.1f'%(float(omstr)), transform=ax.transAxes, fontsize=25, family='bold')
fig.savefig('fig_H_ratio.eps', format='eps')

###

fig, ax = figax(14,8)

ax.plot(Z,Odez, label='$\\rho_{de}(z)/\\rho_c(z)$', lw=lw)
#ax.plot(Z,Omz, label='$\\rho_{m}(z)/\\rho_c(z)$', lw=lw)
ax.plot(Z,RhodeToRhom, label='$\\rho_{de}(z)/\\rho_{m}(z)$', lw=lw)
ax.plot(Z,Qz, label='$q(z)$',  ls='--', lw=lw+2)
ax.plot(Z,QzLCDM, label='$q(z)\\ (\\Lambda \\rm CDM)$',  ls='--', lw=lw+2)
ax.plot(Z,weff, label='$w_{eff}(z)$', ls = ':', lw=lw+3)
setax(ax)
ax.legend(loc='best', frameon=False, fontsize=20)
ax.text(0.85,0.9,'$\\Omega_m$=%.1f'%(float(omstr)), transform=ax.transAxes, fontsize=25, family='bold')
fig.savefig('fig_Quantities.eps', format='eps')
```
